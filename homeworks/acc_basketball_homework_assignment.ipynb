 **1. Reading the CSV File**
- Load the provided CSV file into a pandas DataFrame.
- Display the first 5 rows of the DataFrame to understand the structure of the data.

```python

import pandas as pd

# Load the regular season and playoff data (semicolon-separated and encoding handled)
acc_df = pd.read_csv('acc_players-2324F.csv', header=1)

# Display the first few rows of each dataset
print("Acc Players:")
acc_df.head()



```

**2. Basic Analysis:**).

2a. Calculate the total points scored by all players combined.

```python
total_points = acc_df['PTS'].sum()  # Sum the total points

print(f"Total points for all players: {total_points}")
```

2b. Find the player who has played the most minutes (MP).

```python
max_index = acc_df['MP'].idxmax()

# Retrieve the player's name and minutes played
max_player = acc_df.loc[max_index, 'Player']
minutes = acc_df.loc[max_index, 'MP']

print(f"{max_player} has the most minutes! ({minutes})")
```

2c. Identify the top 5 players in terms of total rebounds (TRB).

```python
top_5_rebound = acc_df.sort_values(by='TRB', ascending=False).head(5)

top_5_rebound
```

**3. Player Filtering:**

3a. Create a new DataFrame containing only players who played more than 500 minutes.

```python
acc_500 = acc_df[acc_df['MP'] > 500]

acc_500.head() # Double-checking if correct.
```

3b. From this filtered DataFrame, determine the player with the highest total assists (AST).

```python
max_index = acc_500['AST'].idxmax()

max_player = acc_500.loc[max_index, 'Player']
assists = acc_500.loc[max_index, 'AST']

print(f"{max_player} has the most assists! ({assists})")
```

3c. Who are the top 3 Assist Leads in the League?

```python
top_3_assists = acc_500.sort_values(by='AST', ascending=False).head(3)

top_3_assists
```

3d. Who are the top 3 Shot Blockers?

```python
# I am assuming BLK is for blockers
top_3_blockers = acc_500.sort_values(by='BLK', ascending=False).head(3)
top_3 = top_3_blockers[['Player', 'BLK']]

print(top_3)
```

**4. School-Based Analysis:**ols

4a. Group the players by School and calculate the total points scored by each school.

```python
school_points_df = acc_df.groupby('School')['PTS'].sum().reset_index()
# kept getting unknown sorting error. Got help from ChatGPT to use reset_index.

school_points = school_points_df.sort_values(by='PTS')

school_points
```

4b. Group the players by School and calculate the total assists (AST) for each team.

```python
school_points_df = acc_df.groupby('School')['AST'].sum().reset_index()
# kept getting unknown sorting error. Got help from ChatGPT to use reset_index.

school_points = school_points_df.sort_values(by='AST')

school_points
```

4c. Sort the schools by total points scored and display the top 3 schools

```python
school_points_df = acc_df.groupby('School')['PTS'].sum().reset_index()

school_points = school_points_df.sort_values(by='PTS', ascending=False)

school_points.head(3)
```
